```python
import pandas as pd
import numpy as np
import os
from tqdm import tqdm

## To keep the plots even after closing the notebook
import plotly.io as pio
pio.renderers.default='notebook'
```

```python
folder_path = 'output/'
# list all files in the directory
files = os.listdir(folder_path)

# filter files that start with 'Take_2' and end with '.csv'
filtered_files = [file for file in files if file.startswith('Take_2') and file.endswith('.csv')]
filtered_files.sort()
# print the list of filtered files
filtered_files
```

```python
len(filtered_files)
```

```python
filename = filtered_files[16]
filename
```

```python


file_path = 'output/'

df_temp_table = pd.read_csv(file_path+filename)

df_temp_table.head(5)
```

```python
## Inspection use only

df_temp_table['data_usable'].value_counts()
```

```python
# Define the time UTC values to exclude
exclude_utc_values = ['21:00', '22:00', '23:00', '00:00', '01:00', '02:00', '03:00', '04:00', '05:00', '06:00',
                      '07:00', '08:00', '09:00', '10:00', '11:00', '12:00', '13:00']

# Filter the dataframe based on the exclusion of time UTC values
filtered_df = df_temp_table[~df_temp_table['Time_UTC'].isin(exclude_utc_values)]

# Check the distribution of values in the 'data_usable' column
distribution = filtered_df['data_usable'].value_counts()

distribution
```

```python
filtered_df = filtered_df.drop('Unnamed: 18', axis=1)
```

```python
filtered_df['event_year'] = filtered_df['Date_UTC'].str[:4]
```

```python
# Filter the dataframe based on the condition
filtered_rows = filtered_df[(filtered_df['data_usable'] == False) & (filtered_df['Date_UTC'].str.startswith('2011'))]

filtered_rows
```

```python
df_unusable_data = filtered_df[filtered_df['data_usable'] == False]

df_unusable_data['event_year'].value_counts()
```

```python
df_unusable_data.to_csv('output/unusable_data_rev_2.csv', index=False)
```
