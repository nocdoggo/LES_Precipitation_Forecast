```python
bc = pd.read_csv('04_binned_lat_lon_opencv_12f13s_lake_0.csv')
bc
```

```python
h_list = []
m_list = []
for t in bc['Time']:
    h_list.append(int(t[1:3]))
    m_list.append(int(t[-2:]))
bc['Hour'] = h_list
bc['Min'] = m_list
bc
```

```python
y_list = []
mon_list = []
day_list = []
for d in bc['Date']:
    d = str(d)
    y_list.append(int(d[:4]))
    mon_list.append(int(d[4:6]))
    day_list.append(int(d[6:]))
bc['Year'] = y_list
bc['Month'] = mon_list
bc['Day'] = day_list
bc
```

```python
print(int(bc.iloc[1, 1][1:]))
```

```python
nd_list = []
nt_list = []
for i in range(len(bc)):
    print(bc.iloc[i, 18], bc.iloc[i, 19], bc.iloc[i, 20], bc.iloc[i, 16], bc.iloc[i, 17])
    my_datetime = datetime(bc.iloc[i, 18], bc.iloc[i, 19], bc.iloc[i, 20], bc.iloc[i, 16], bc.iloc[i, 17], tzinfo = pytz.utc)
    #my_datetime_utc = my_datetime.strftime('%Y-%m-%d %H:%M:%S %Z%z')
    #print(my_datetime_utc)
    my_datetime_cst = my_datetime.astimezone(pytz.timezone('Etc/GMT+6')).strftime('%Y%m%d %H%M')
#     print(my_datetime_cst)
#     print(my_datetime_cst[:8])
#     print(my_datetime_cst[-4:])
    #print(type(my_datetime_cst))
    nd_list.append(my_datetime_cst[:8])
    nt_list.append(my_datetime_cst[-4:])
    
```

```python
bc['CST_Date'] = nd_list
bc['CST_Time'] = nt_list
```
