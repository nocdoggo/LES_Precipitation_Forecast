```python
import math
from datetime import datetime, timedelta

def round_datetime_to_nearest_hour(obj_arr, STATE_ID):
    arr_len = len(obj_arr)
    Date_CST = []
    Time_CST = []
    for i in range(arr_len):
        # iterate through
        date_item = obj_arr['Date'][i]
        time_item = obj_arr['Time'][i]
        t = datetime.strptime(date_item + " " + time_item, "%Y-%m-%d %H:%M")
        # Calculate the number of minutes past the last full hour
        minutes_past_hour = t.minute + t.second / 60
        # Round up to the next whole number of hours if the time is more than 30 minutes past the hour,
        # or round down to the current hour if it's less than 30 minutes past
        if minutes_past_hour >= 30:
            num_hours = math.ceil(minutes_past_hour / 60)
        else:
            num_hours = 0
        # Create a new datetime object representing the rounded time
        if STATE_ID in ['IL', 'WI']:
            # No need to dial back one hour
            rounded_time_temp = t + timedelta(hours=num_hours)
        else:
            rounded_time_temp = t + timedelta(hours=num_hours-1)
        rounded_time=datetime(year=rounded_time_temp.year,
                         month=rounded_time_temp.month,
                         day=rounded_time_temp.day,
                         hour=rounded_time_temp.hour, minute=0, second=0)
        result_stamp = rounded_time.strftime("%Y%m%d %H:%M")
        
        new_date, new_time = result_stamp.split(' ')
        Date_CST.append(new_date)
        Time_CST.append(new_time)
        
    obj_arr['Date_CST'] = Date_CST
    obj_arr['Time_CST'] = Time_CST
    return obj_arr
```

### Try to gather the LES cloud coverage information

One of the challenges here is that the duration of the cloud coverage can run into multiple days. However, due to the use of Band 1, aka visible wavelength band, we will need to figure out how to count the data from multiple days.

The defined visible daytime range is from `1400` to `2100`, a timespan of 8 hours per day.

```python
import datetime

end_date_list = []
end_time_list = []

for i in range(len(df_obs_snow_matched)):
    start_date = datetime.datetime.strptime(str(df_obs_snow_matched.loc[i, 'LES_Cloud_Cover_Start_Date']), '%Y%m%d')
    start_time = datetime.datetime.strptime(str(df_obs_snow_matched.loc[i, 'LES_Cloud_Cover_Start_Time']), '%H%M')

    duration_hours = df_obs_snow_matched.loc[i, 'Duration of LES Cloud']

    # Calculate the number of days and hours to add
    num_days = int(duration_hours // 8)
    num_hours = int(duration_hours % 8)
#     print(num_days)
#     print(num_hours)

    # Calculate the end datetime
    end_date_time = start_date + timedelta(days=num_days) + timedelta(hours=num_hours)
    end_time_hour = start_time.hour + num_hours + ((num_hours) // 8) * 3
    end_time_minute = start_time.minute
    end_time_second = start_time.second

#     # Calculate the end time
#     if end_time_minute >= 60:
#         end_time_hour += 1
#         end_time_minute -= 60

    end_time_list.append(f"{end_time_hour:02}{end_time_minute:02}")
    end_date_list.append(end_date_time.strftime('%Y%m%d'))

print(end_date_list)
print(end_time_list)
```

**Construct function to read the lake data.**

```python
import numpy as np

def lake_basic_stat(filedir):
    df_lake_raw = pd.read_csv(filedir)
    
    # Acquire the data from the specific columns
    x, y, i = df_lake_raw.longitude, df_lake_raw.latitude, df_lake_raw.value
    
    # Intermediate stats
    mean_i = np.nanmean(i)
    std_i = np.nanstd(i)
    p75 = np.nanpercentile(i, 25)
    
    cut_off_threshold = mean_i - 1*std_i
    
#     print(cut_off_threshold)
    
    if cut_off_threshold <= 0:
        cut_off_threshold = p75
        
#     print(cut_off_threshold)
    
    count = len([x for x in i if x > cut_off_threshold and not math.isnan(x)])
    total_intensity_unweighted = np.nansum(i)
    normalized_i = [x/total_intensity_unweighted for x in i if not np.isnan(x)]
    cut_off_threshold_normalized = cut_off_threshold / total_intensity_unweighted
    total_intensity_normalized = sum(val for val in normalized_i if val > cut_off_threshold_normalized)
    
    percent_cloud_cover = count / len(i) * 100
    return count, total_intensity_unweighted, total_intensity_normalized, percent_cloud_cover, cut_off_threshold
```
