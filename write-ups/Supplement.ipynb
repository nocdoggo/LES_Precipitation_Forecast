```python
# Read the CSV file into df_info_new
df_info_new = pd.read_csv('results.csv')

# Reconstruct the text
reconstructed_text = ''
for _, row in df_info_new.iterrows():
    epoch_string = f"Epoch {int(row['Epoch'])}/60\n"
    metrics_string = f"106/106 [==============================] - {int(row['Seconds'])}s 1s/step - "
    metrics_string += f"loss: {row['Loss']} - accuracy: {row['Accuracy']} - "
    metrics_string += f"val_loss: {row['Val_Loss']} - val_accuracy: {row['Val_Accuracy']} - "
    metrics_string += f"lr: {row['Lr']}\n"
    
    reconstructed_text += epoch_string + metrics_string
    
# Define a function to format numbers to 4 decimal places
def format_match(match):
    return "{:.4f}".format(float(match.group()))

# Use regex to find and replace numbers related to each metric
for metric in ['loss', 'accuracy', 'val_loss', 'val_accuracy']:
    pattern = "(?<=" + metric + r":\s)\d+\.\d+"
    reconstructed_text = re.sub(pattern, format_match, reconstructed_text)

reconstructed_text = reconstructed_text.replace('0.001', '0.0010')
reconstructed_text = re.sub(r'(\d)e-(\d+)', r'\1.0000e-\2', reconstructed_text)

print(reconstructed_text)
```
